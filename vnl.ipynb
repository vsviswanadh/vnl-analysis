```python
numeric_cols=df.select_dtypes(include=["int","float"]).columns
corr_matrix=df[numeric_cols].corr()
corr_matrix
```

```python
plt.figure(figsize=(12,8))
sb.heatmap(corr_matrix,annot=True,cmap="coolwarm",linewidths=2)
plt.title("correlation matrix heatmap")
plt.show()
```

```python
position_counts=df["Position"].value_counts()
position_counts
```

```python
plt.pie(position_counts,labels=position_counts.index,autopct="%1.1f%%",startangle=90)
plt.title("distribution of players position")
plt.show()
```

```python
avg_attack_by_country=df.groupby("Country")["Attack"].mean()
avg_attack_by_country.sort_values(ascending=False)
```

```python
avg_attack_by_country.sort_values(ascending=False).plot(kind="bar")
plt.title("average attack by each country")
plt.xlabel("country")
plt.ylabel("average attack")
plt.show()
```

```python
average_attack_by_position=df.groupby("Position")["Attack"].mean()
average_attack_by_position.sort_values(ascending=False).plot(kind="bar",color="purple",edgecolor="black")
plt.title("average by position")
plt.xlabel("position")
plt.ylabel("averagea attack score")
plt.show()
```

```python
serve_trend_by_age=df.groupby("Age")["Serve"].mean()
serve_trend_by_age.plot(kind="line",marker=".",linestyle="-",color="red")
plt.title("serve trend by age")
plt.xlabel("age")
plt.ylabel("serve")
plt.show()
```

```python
total_attack_block_by_country=df.groupby("Country")[["Attack","Block"]].sum()
total_attack_block_by_country.sort_values(ascending=False,by="Attack").plot(kind="bar",stacked=True,colormap="viridis")
plt.title("total attack and block by each country")
plt.xlabel("country")
plt.ylabel("attack,block")
plt.show()
```
